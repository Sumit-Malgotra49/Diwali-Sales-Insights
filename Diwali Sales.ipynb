```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing the csv file
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
df
```

```python
#getting the basic info
df.info()
```

```python
#dropping unwanted columns
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
df.info()
```

```python
#checking the null values
pd.isnull(df).sum()
```

```python
#dropping the empty values
df.dropna(subset=['Amount'], inplace=True)
df.info()
```

```python
#checking empty values again
pd.isnull(df).sum()
```

```python
#Getting the description
df.describe()
```

```python
#changing the type of Amount
df['Amount'] = df['Amount'].astype('int')
#checking the info again
df.info()
```

# **Exploratory Data Analysis**


```python
# plotting bar chart for Gender count

ax = sns.countplot(x = 'Gender', data = df)
for bars in ax.containers:
  ax.bar_label(bars)
```

We can clearly see that females are shopping more than males during the Diwali season

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

Here, you can also see that females are spending more than males during the Diwali sales.

```python
# Gender by Age Group
ax = sns.countplot(x = 'Age Group', data = df, hue = 'Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
# Total Amount vs Age Group

sales_gen = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_gen)
```

The graphs above indicate that the majority of buyers are females aged between 26 and 35 years

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(5)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
```

```python
ax = sns.countplot(x = 'Marital_Status', data = df)
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

The graphs above show that the majority of buyers are married women, and they possess high purchasing power

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sector_sales = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False ).head(5)
sns.barplot(x='Occupation', y= 'Amount', data =sector_sales)
sns.set(rc={'figure.figsize':(7,5)})
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Orders')
```
